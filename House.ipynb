```python
import turtle
turtle.bgcolor("light blue")
t = turtle.Turtle()
t.speed(3)
t.pensize(5)
wall_length = 200
wall_height = 220
door_length = 80
door_height = 125
wall_length_1 = 150
wall_height_1 = 220
tree_length = 40
tree_height = 100
triangle = 120
triangle_1 = 100
triangle_2 = 80
triangle_side = triangle * 2**0.5
triangle_side_1 = triangle_1 * 2**0.5
triangle_side_2 = triangle_2 * 2**0.5


t.penup()
t.goto(-500,-300)
t.pendown()

t.fillcolor("blue")
t.begin_fill()
for i in range(2):
    t.fd(wall_length)
    t.lt(90)
    t.fd(wall_height)
    t.lt(90)
t.end_fill()

t.penup()
t.fd(60)
t.pendown()
t.pensize(5)

t.fillcolor("light green")
t.begin_fill()
for i in range(2):
    t.fd(door_length)
    t.lt(90)
    t.fd(door_height)
    t.lt(90)
t.end_fill()

t.penup()
t.fd(door_length + 60)
t.pendown()

t.fillcolor("yellow")
t.begin_fill()
t.lt(30)
t.fd(wall_length_1)
t.lt(60)
t.fd(wall_height_1)
t.lt(120)
t.fd(wall_length_1)
t.end_fill()


t.penup()
t.rt(90)
t.pendown()

t.fillcolor("violet")
t.begin_fill()
t.fd(wall_length)
t.lt(120)
t.fd(wall_length)
t.end_fill()

t.penup()
t.bk(wall_length)
t.lt(180)
t.pendown()

t.fillcolor("orange")
t.begin_fill()
t.rt(30)
t.fd(wall_length_1)
t.rt(90)
t.fd(wall_length)
t.rt(90)
t.fd(wall_length_1)
t.end_fill()

t.penup()
t.lt(90)
t.fd(50)
t.lt(90)
t.fd(25)
t.pendown()

t.fillcolor("red")
t.begin_fill()
t.fd(50)
t.rt(120)
t.fd(50)
t.rt(60)
t.fd(50)
t.rt(120)
t.fd(50)
t.end_fill()

t.penup()
t.home()
t.goto(200,350)
t.pendown()

t.fillcolor("yellow")
t.begin_fill()
t.circle(50)
t.end_fill()

t.penup()
t.lt(90)
t.fd(50)
t.rt(90)
t.pendown()
t.pensize(3)

for i in range(4):
        t.penup()
        t.fd(50)
        t.pendown()
        t.fd(80)
        t.penup()
        t.bk(80 + 50)
        t.lt(90)

t.pensize(2)
t.lt(45)

for i in range(4):
        t.penup()
        t.fd(50)
        t.pendown()
        t.fd(50)
        t.penup()
        t.bk(50 + 50)
        t.lt(90)

t.penup()
t.home()
t.goto(500,-400)
t.pensize(5)
t.pendown()

t.fillcolor("brown")
t.begin_fill()
for i in range(2):
    t.fd(tree_length)
    t.lt(90)
    t.fd(tree_height)
    t.lt(90)
t.end_fill()

t.penup()
t.lt(90)
t.fd(tree_height)
t.rt(90)
t.fd(tree_length / 2)
t.pendown()

t.fillcolor("green")
t.begin_fill()
t.fd(triangle)
t.lt(135)
t.fd(triangle_side)
t.lt(90)
t.fd(triangle_side)
t.lt(135)
t.fd(triangle)
t.end_fill()

t.penup()
t.lt(90)
t.fd(triangle)
t.rt(90)
t.pendown()

t.fillcolor("green")
t.begin_fill()
t.fd(triangle_1)
t.lt(135)
t.fd(triangle_side_1)
t.lt(90)
t.fd(triangle_side_1)
t.lt(135)
t.fd(triangle_1)
t.end_fill()

t.penup()
t.lt(90)
t.fd(triangle_1)
t.rt(90)
t.pendown()

t.fillcolor("green")
t.begin_fill()
t.fd(triangle_2)
t.lt(135)
t.fd(triangle_side_2)
t.lt(90)
t.fd(triangle_side_2)
t.lt(135)
t.fd(triangle_2)
t.end_fill()


```
